### Exercise 1

Proving the statement is equivalent to proving that for every $A \in \mathcal{M}(n\times n)$ we have $$||A||^2 = \rho(A'A).$$ 

First note that $||A||^2 = \left(\sup_{x\in \mathbb{R}^n \setminus \{0\}} \frac{||Ax||}{||x||} \right)^2 = \left(\sup_{x\in \mathbb{R}^n \setminus \{0\}, ||x||=1} ||Ax|| \right)^2$, where the second equality is w.l.o.g. (it should be obvious from the context when the spectral norm and vector norms like the Euclidean norms are used). The function $x \mapsto ||Ax||$ is continuous by continuity of a linear operator and of the norm, and the set $\left\lbrace x\in \mathbb{R}^n \setminus \{0\}, ||x||=1 \right\rbrace$ is compact, so by the theorem of the maximum the maximum is attained, and we can replace the sup by a max, $$\max_{x\in \mathbb{R}^n \setminus \{0\}, ||x||=1} ||Ax|| = \max_{x\in \mathbb{R}^n \setminus \{0\}, ||x||=1} \left[x'(A'A)x \right]^{1/2} = \max_{x\in \mathbb{R}^n \setminus \{0\}, \sum_{i=1}^nx_i^2=1} \left[x'(A'A)x \right].$$  

Then we solve the maximization problem using the Lagrange method for constrained optimization: \begin{eqnarray*}
&& \mathcal{L} = x'(A'A)x  + \lambda \left[1-\sum_{i=1}^nx_i^2\right] \\
& \Rightarrow & \frac{\partial \mathcal{L}}{x} =  (A'A)x=\lambda x
\end{eqnarray*}
To conclude, note that by definition of the spectral radius, $$\rho(A'A)=\max\left\lbrace |\lambda|: \exists x\in \mathbb{R}\setminus \{0\}, A'Ax = \lambda x \right\rbrace.$$
Note that the FOC implies $x'(A'A)x=\lambda x'x = \lambda ||x||^2 = \lambda$, since $||x||=1$. 

So $\left(\max_x \frac{||Ax||}{||x||}\right)^2 = \lambda = |\lambda|$, where the last equality follows from $A'A$ being positive semi-definitive (quadratic form), and thus with non-negative eigenvalues. This proves the claim. 

### Exercise 2


We have $\rho(A)<1$. By Gelfand's formula, $\lim_{k\to\infty} ||A^k||^{1/k} = \rho(A)$, i.e. $\forall \epsilon >0$ $\exists K \in \mathbb{R}$ s.t. $\forall k \geq K$, $\left| ||A^k||^{1/k}-\rho(A) \right| \leq \epsilon $. 

This implies $||A^k||^{1/k}-\rho(A) \leq \epsilon$, i.e. $||A^k||^{1/k} \leq \rho(A)+\epsilon$, i.e. $||A^k|| \leq (\rho(A)+\epsilon)^k$. Denote $r=\rho(A)+\epsilon < 1$ because $\epsilon$ can be chosen arbitrarily small and $\rho(A)<1$. Also denote $C=1\in \mathbb{N}$. Thus we have an upper bound for the high end of the sequence of spectral norms. 

Now let's consider the low end of this sequence. For any $k'<K$, we have $||A^{k'}|| \leq \max_{k''=1,...,K}||A^{k''}|| $. Now, $r<1$ implies that for any such $k'$, $||A^{k'}|| \leq \max_{k''=1,...,K}||A^{k''}|| \leq \max_{k''=1,...,K}||A^{k''}|| \frac{r^{k'}}{r^K}$. That is, $||A^{k'}|| \leq \frac{\max_{k''=1,...,K}||A^{k''}||}{r^K}r^{k'}$. Thus we have also bounded the lower end of the sequence. 

It suffices to note that the bound on the higher end of the sequence (for $k\geq K$) implies that $C \equiv \frac{\max_{k''=1,...,K}||A^{k''}||}{r^K} \geq 1$, to conclude that we can also bound the high-end of the sequence using $C$. 

This completes the proof of the statement that if $\rho(A)<1$, there exists $r=\rho(A)+\epsilon<1$ and $C=\frac{\max_{k''=1,...,K}||A^{k''}||}{r^K}$ such that $||A^k|| \leq r^k C$ (if we want $C\in \mathbb{N}$ we can take the smallest integer greater than $C$). (Note that $r$ and $C$ depend on each other, through $\epsilon$ and $K$). 

This last statement implies that $\sum_{k=0}^\infty ||A^k|| \leq C\sum_{k=0}^\infty r^k = \frac{C}{1-r} < \infty$.

### Exercise 6

Define the model parameters used for testing the conmputations.

```julia
A = [.8 -.1;
-.1 .8]
C = eye(2)
D = eye(2)
beta = .9
```

Define a function taking as argument the primitives of the problem, which returns $P$ and $\delta$. It also tests the stability condition in the Lyapunv equation. 

The price function is then defined as $$p^*(x) = x'P^*x + \delta^*.$$

It is computed for the primitives given above and $u_t \sim \mathcal(N)(0,1)$ iid, and prices are simulated over a $T=100$ horizon and plotted. 

```julia
function compute_price(beta::Float64, A::Array{Float64,2}, C::Array{Float64,2}, D::Array{Float64,2})
    # solve the discrete Lyapunov equation
    M = beta*A'*D*A
    Lambda = sqrt(beta)*A'
    if maximum(abs(eig(Lambda)[1])) >= 1/sqrt(beta) # stability condition rho(Lambda)<1/sqrt(beta)
        error("ARGUMENT ERROR: A, beta: Stability condition rho(A) < 1/sqrt(beta) violated.")
    else
        P0 = ones(2,2)*.9
        distance = 1
        conv_crit = 1e-3
        maxit = 100
        iter = 0
        while distance > conv_crit || iter > maxit
            P1 = Lambda*P0*Lambda' + M
            distance = norm(P1-P0, Inf)
            P0 = deepcopy(P1)
            #     println("distance"); println(distance) # print distance to check convergence
            iter = iter +1
        end
        P = P0
        delta = beta/(1-beta) * trace(C'*(D+P)*C)
    end
    return P, delta
end
```

```julia
# %matplotlib inline
using PyPlot

P, delta = compute_price(beta, A, C, D)

T = 100
# state x process N(0,1)
w = randn(2, T+1)
x = zeros(2, T+1)
for t = 1:T
    x[:,t+1] = A*x[:,t] + C*w[:,t+1]
end
# simulate prices 
p = Array(Float64,T+1)
for t = 1:T+1
    p[t] = (x[:,t]'*P*x[:,t])[1] + delta
end

fig, ax = subplots()
ax[:plot](collect(1:T), p[2:T+1], linewidth=2, alpha=0.6)
ax[:set_xlabel](L"$t$")
ax[:set_ylabel](L"$p_t$")
ax[:legend](loc="upper right")
```

Now let's make the contour plot. For that use the QuantEcon package. I obtain the same graph as in the problem set. 

```julia
using QuantEcon

# == Set up grid for plotting == #
x_grid = linspace(-10,10, 200)
y_grid = linspace(-10, 10, 200)
X, Y = meshgrid(x_grid, y_grid)

function gen_prices(P, delta)
    price = Array(Float64, length(x_grid), length(y_grid))
    for i in 1:length(x_grid)
        for j in 1:length(y_grid)
            state = [x_grid[i];y_grid[j]]
            price[i,j] = (state'*P*state + delta)[1]
        end
    end
    return price
end

# helper function to prepare axis
function prep_ax()
    fig, ax = subplots()
    ax[:xaxis][:grid](true, zorder=0)
    ax[:yaxis][:grid](true, zorder=0)
    return ax
end

function plot1()
    ax = prep_ax()
    Z = gen_prices(P, delta)
    ax[:contourf](X, Y, Z, 8, alpha=0.6, cmap=ColorMap("jet"))
    cs = ax[:contour](X, Y, Z, 8, colors="black")
    ax[:clabel](cs, inline=1, fontsize=10)
end
plot1()
```
